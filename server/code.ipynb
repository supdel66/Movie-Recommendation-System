```python
import pandas as pd
import numpy as np
```

```python
import difflib #name of movie from user input
#if spelling mistakes here then we need to compare and give the movie name with best match

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df=pd.read_csv('movies.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.fillna(' ',inplace=True)
df.isnull().sum()
```

```python
selected_features=['genres','keywords','overview','tagline','cast','director']
df['combined']=' '
for feature in selected_features:
    df['combined']=df['combined']+' '+df[feature]
df['combined']    
     
```

```python
#converting to numberical values

feature_extraction=TfidfVectorizer(min_df=1, stop_words='english',lowercase=True)
featurevectors=feature_extraction.fit_transform(df['combined'])
```

```python
print(featurevectors)
```

```python
similarity=cosine_similarity(featurevectors)
print(similarity)
```

cosine similarity is cosine of angle between the document lines suppose iphone and samsung count lines 

```python
similarity.shape
```

```python
#getting movie name
movie_name=input('enter the name of movie')
```

```python
list_of_all_title=df['original_title'].values
print(list_of_all_title)
```

```python
find_close_match=difflib.get_close_matches(movie_name,list_of_all_title)
#suppose the input movie is not in my dataset then?
if find_close_match:
    print(find_close_match)
    close_match=find_close_match[0]
else:
    print('Sorry the movie is not in my training dataset. I shall update it as soon as possible. Thank You!')

```

```python
a=difflib.ndiff(movie_name,list_of_all_title[0])
print('\n'.join(a))
```

```python
movie_index=df[df['original_title']==close_match]['index'].values[0]  
#values le array ma lagxa [0] le just value matra dinxa 
#.values garena bhane chai dataframe dinxa
print(movie_index)
```

```python
similarity_score=list(enumerate(similarity[movie_index]))
print(similarity_score)
```

0, 0.233  0 is index and 0.233 is similarity with the movie index given 

```python
len(similarity_score)
```

```python
#sorting
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

x  is similarity score  x[1] is  the similarity score that is to be sorted 

```python
print('Your Input Movie:',movie_name)
print('Movies Suggested for you:\n')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df['index']==index]['original_title'].values[0]
    if(i<30):
        print(i,' ',title_from_index)
        i+=1

```
